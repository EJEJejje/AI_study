```python
h,m,s = map(int, input().split())
timer = int(input())
```

```python
s += timer
m_add, s =divmod(s,60)
m += m_add

h_add, m = divmod(m,60)
h += h_add

if h >= 24:
    h -= 24
print(f'{h},{m},{s}')
```

```python
H, M, S = map(int, input().split())
D = int(input()) 

S += D % 60 #200일때 20추가
D = D // 60 # +1
if S >= 60:
    S -= 60 #초기화
    M += 1 #+ 1

M += D % 60
D = D // 60
if M >= 60:
    M -= 60
    H += 1

H += D % 24
if H >= 24:
    H -= 24

print(H,M,S)
```
